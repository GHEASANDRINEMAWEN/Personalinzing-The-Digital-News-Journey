```python

# FILE: history.parquet
# PURPOSE:
#   1. Flatten user-level lists into event-level rows
#   2. Clean irregular list lengths (no data loss)
#   3. Perform EDA on scroll % and read time


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pyarrow.parquet as pq


file_path = "Downloads/ebnerd_demo/train/history.parquet"
history = pq.read_table(file_path).to_pandas()
print("Initial shape:", history.shape)
print("Columns:", history.columns.tolist())


def max_len(row):
    lengths = [len(x) for x in [
        row.get('impression_time_fixed'),
        row.get('scroll_percentage_fixed'),
        row.get('article_id_fixed'),
        row.get('read_time_fixed')
    ] if isinstance(x, (list, np.ndarray))]
    return max(lengths) if lengths else 0

history['max_len'] = history.apply(max_len, axis=1)

# ============================================================
# STEP 3️⃣: PAD SHORTER LISTS WITH NaN (no data loss)
# ============================================================
list_cols = ['impression_time_fixed', 'scroll_percentage_fixed',
             'article_id_fixed', 'read_time_fixed']

def ensure_list(x):
    """Convert any value to a list"""
    if isinstance(x, np.ndarray):
        return x.tolist()
    if isinstance(x, list):
        return x
    if x is None or (isinstance(x, float) and np.isnan(x)):
        return []
    return [x]

# Convert all columns to lists first
for col in list_cols:
    history[col] = history[col].apply(ensure_list)

# Pad each column to match max_len for that row
for col in list_cols:
    history[col] = history.apply(
        lambda row: row[col] + [None] * (row['max_len'] - len(row[col])),
        axis=1
    )

# Verify all lists have the same length per row
print("\n Verifying list lengths...")
for idx, row in history.iterrows():
    lengths = [len(row[col]) for col in list_cols]
    if len(set(lengths)) > 1:
        print(f" Row {idx} has mismatched lengths: {lengths}")


history_flat = history.explode(list_cols, ignore_index=True)

history_flat['impression_time_fixed'] = pd.to_datetime(
    history_flat['impression_time_fixed'], errors='coerce'
)
history_flat['scroll_percentage_fixed'] = pd.to_numeric(
    history_flat['scroll_percentage_fixed'], errors='coerce'
)
history_flat['read_time_fixed'] = pd.to_numeric(
    history_flat['read_time_fixed'], errors='coerce'
)
history_flat['article_id_fixed'] = pd.to_numeric(
    history_flat['article_id_fixed'], errors='coerce'
)

# Remove rows where both scroll and read_time are missing
history_flat = history_flat.dropna(
    subset=['scroll_percentage_fixed', 'read_time_fixed'], how='all'
)

print("\n Flattened dataset ready!")
print("Shape after flattening:", history_flat.shape)
print(history_flat.head())

```

```python

# FILE: behaviors.parquet
# PURPOSE:
#   1. Flatten impression-level data into article-level rows
#   2. Create a binary 'clicked' label for each candidate
#   3. Explore device usage, click rates, and impression stats
#
# TEAM: Sandrine | Amir | Maxine
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pyarrow.parquet as pq


file_path = "Downloads/ebnerd_demo/train/behaviors.parquet"
behaviors = pq.read_table(file_path).to_pandas()

print("Initial shape:", behaviors.shape)
print("Columns:", behaviors.columns.tolist())
print("\nPreview:")
print(behaviors.head(2))

list_cols = ["article_ids_inview", "article_ids_clicked"]

def ensure_list(x):
    """Ensure all entries are lists (even empty ones)"""
    if isinstance(x, list):
        return x
    elif isinstance(x, np.ndarray):
        return x.tolist()
    elif pd.isna(x):
        return []
    elif isinstance(x, str):
        # handle accidental string representations of lists
        try:
            return eval(x)
        except:
            return [x]
    else:
        return [x]

for col in list_cols:
    behaviors[col] = behaviors[col].apply(ensure_list)

# Each impression becomes multiple rows (one per article candidate)
behaviors_flat = behaviors.explode("article_ids_inview", ignore_index=True)


# Mark which candidate was clicked (1) or not (0)
behaviors_flat["clicked"] = behaviors_flat.apply(
    lambda row: 1 if row["article_ids_inview"] in row["article_ids_clicked"] else 0,
    axis=1
)

print("\n Flattened dataset ready!")
print("Shape after flattening:", behaviors_flat.shape)
print(behaviors_flat.head(20))


```

```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt

# -----------------------------------------------------------------------------
# 1. LOCATION AND CONSTANTS
# -----------------------------------------------------------------------------
lat = math.radians(4.05)       # Douala latitude (N positive)
lon = 9.7                      # Longitude (E positive)
timezone = 1                   # UTC+1 (Cameroon)
beta = abs(math.degrees(lat))  # Tilt = latitude
gamma = 0                      # Azimuth (0 = facing equator)
albedo = 0.2                   # Ground reflectivity
G_sc = 1367                    # Solar constant (W/m²)
hours = np.arange(0, 24, 1)    # Local time 0–23
days = [80, 172]               # March 21, June 21 (2025)
day_names = ['March_21', 'June_21']

# -----------------------------------------------------------------------------
# 2. STORAGE ARRAYS
# -----------------------------------------------------------------------------
solar_altitude = np.zeros((len(days), len(hours)))
solar_azimuth = np.zeros((len(days), len(hours)))
extraterrestrial = np.zeros((len(days), len(hours)))
clear_sky_horizontal = np.zeros((len(days), len(hours)))
beam_horizontal = np.zeros((len(days), len(hours)))
diffuse_horizontal = np.zeros((len(days), len(hours)))
beam_tilted = np.zeros((len(days), len(hours)))
diffuse_tilted = np.zeros((len(days), len(hours)))
reflected_tilted = np.zeros((len(days), len(hours)))
total_tilted = np.zeros((len(days), len(hours)))

# -----------------------------------------------------------------------------
# 3. FUNCTIONS
# -----------------------------------------------------------------------------
def sind(x): return math.sin(math.radians(x))
def cosd(x): return math.cos(math.radians(x))
def tand(x): return math.tan(math.radians(x))
def asind(x): return math.degrees(math.asin(x))
def atan2d(y, x): return math.degrees(math.atan2(y, x))

# -----------------------------------------------------------------------------
# 4. MAIN LOOP OVER DAYS
# -----------------------------------------------------------------------------
for d, day in enumerate(days):
    delta = 23.45 * sind(360 * (284 + day) / 365)
    B = 360 * (day - 81) / 364
    EoT = 9.87 * sind(2 * B) - 7.53 * cosd(B) - 1.5 * sind(B)
    LSTM = 15 * timezone
    TC = 4 * (lon - LSTM) + EoT
    S_ex = G_sc * (1 + 0.033 * cosd(360 * day / 365))

    for h, hour in enumerate(hours):
        solar_time = hour + TC / 60
        hour_angle = 15 * (solar_time - 12)

        sin_hs = sind(math.degrees(lat)) * sind(delta) + cosd(math.degrees(lat)) * cosd(delta) * cosd(hour_angle)
        sin_hs = max(min(sin_hs, 1), -1)
        hs = asind(sin_hs)
        solar_altitude[d, h] = hs

        if hs <= 0:
            continue

        # Extraterrestrial radiation
        Gex = S_ex * sind(hs)
        extraterrestrial[d, h] = Gex

        # Clear sky radiation
        air_mass = 1 / max(sind(hs), 1e-6)
        trans = 0.7 ** (air_mass ** 0.678)
        G_clear = trans * Gex
        clear_sky_horizontal[d, h] = G_clear

        # Beam and diffuse split
        G_beam = 0.9 * G_clear
        G_diff = 0.1 * G_clear
        beam_horizontal[d, h] = G_beam
        diffuse_horizontal[d, h] = G_diff

        # Tilted surface
        cos_theta = sind(delta) * sind(math.degrees(lat) - beta) + cosd(delta) * cosd(math.degrees(lat) - beta) * cosd(hour_angle)
        cos_theta = max(cos_theta, 0)
        RB = cos_theta / max(sind(hs), 1e-6)
        RD = (1 + cosd(beta)) / 2
        RR = (1 - cosd(beta)) / 2

        G_beam_tilt = RB * G_beam
        G_diff_tilt = RD * G_diff
        G_refl_tilt = RR * albedo * G_clear
        G_total_tilt = G_beam_tilt + G_diff_tilt + G_refl_tilt

        beam_tilted[d, h] = G_beam_tilt
        diffuse_tilted[d, h] = G_diff_tilt
        reflected_tilted[d, h] = G_refl_tilt
        total_tilted[d, h] = G_total_tilt

        # Solar azimuth
        num = sind(hour_angle)
        den = cosd(hour_angle) * sind(math.degrees(lat)) - tand(delta) * cosd(math.degrees(lat))
        az = atan2d(num, den)
        solar_azimuth[d, h] = (az + 360) % 360

# -----------------------------------------------------------------------------
# 5. EXPORT TO EXCEL
# -----------------------------------------------------------------------------
excel_file = 'Douala_Solar_All.xlsx'
with pd.ExcelWriter(excel_file, engine='xlsxwriter') as writer:
    for d, name in enumerate(day_names):
        df = pd.DataFrame({
            'Hour': hours,
            'SolarAltitude_deg': solar_altitude[d],
            'SolarAzimuth_deg': solar_azimuth[d],
            'Extraterrestrial_Wm2': extraterrestrial[d],
            'ClearSkyHorizontal_Wm2': clear_sky_horizontal[d],
            'BeamTilted_Wm2': beam_tilted[d],
            'DiffuseTilted_Wm2': diffuse_tilted[d],
            'ReflectedTilted_Wm2': reflected_tilted[d],
            'TotalTilted_Wm2': total_tilted[d]
        })
        df.to_excel(writer, sheet_name=name, index=False)

print(f"Excel file created: {excel_file}")

# -----------------------------------------------------------------------------
# 6. PLOTTING
# -----------------------------------------------------------------------------
plt.figure(figsize=(10,5))
plt.plot(hours, extraterrestrial[0], 'r-', label='Extraterrestrial (Mar)')
plt.plot(hours, clear_sky_horizontal[0], 'r--', label='Clear-sky (Mar)')
plt.plot(hours, extraterrestrial[1], 'b-', label='Extraterrestrial (Jun)')
plt.plot(hours, clear_sky_horizontal[1], 'b--', label='Clear-sky (Jun)')
plt.xlabel('Hour'); plt.ylabel('Irradiance (W/m²)')
plt.title('Extraterrestrial & Clear-Sky Radiation - Douala')
plt.legend(); plt.grid(True)
plt.tight_layout()
plt.savefig('Douala_Extraterrestrial_and_ClearSky.png')

plt.figure(figsize=(10,5))
plt.plot(hours, clear_sky_horizontal[0], 'r-o', label='Horizontal (Mar)')
plt.plot(hours, total_tilted[0], 'r--s', label='Tilted (Mar)')
plt.plot(hours, clear_sky_horizontal[1], 'b-o', label='Horizontal (Jun)')
plt.plot(hours, total_tilted[1], 'b--s', label='Tilted (Jun)')
plt.xlabel('Hour'); plt.ylabel('Irradiance (W/m²)')
plt.title(f'Horizontal vs Tilted (β={beta:.1f}°) - Douala')
plt.legend(); plt.grid(True)
plt.tight_layout()
plt.savefig('Douala_Horizontal_vs_Tilted.png')

plt.show()
```

```python
import math

# --- Constants and Inputs ---
Gsc = 1367             # solar constant (W/m²)
lat = 4.05             # Douala latitude (°N)
lon = 9.7              # Douala longitude (°E)
timezone = +1          # UTC offset
albedo = 0.2
h = 1.0                # object height (m)
day = 32               # Feb 1
LT = 8.0               # local time (hours)

# --- 1. Equation of Time (EoT) ---
B = (360/365)*(day - 81)
EoT = 9.87*math.sin(math.radians(2*B)) - 7.53*math.cos(math.radians(B)) - 1.5*math.sin(math.radians(B))

# --- 2. Local Standard Time Meridian (LSTM) ---
LSTM = 15 * timezone

# --- 3. Time Correction (TC) ---
TC = 4*(lon - LSTM) + EoT   # minutes

# --- 4. Local Solar Time (LST) ---
LST = LT + TC/60

# --- 5. Hour Angle (H) ---
H = 15*(LST - 12)   # degrees

# --- 6. Declination (δ) ---
delta = 23.45 * math.sin(math.radians((360/365)*(284 + day)))

# --- 7. Solar Altitude (α) ---
sin_alpha = math.sin(math.radians(lat))*math.sin(math.radians(delta)) + \
            math.cos(math.radians(lat))*math.cos(math.radians(delta))*math.cos(math.radians(H))
alpha = math.degrees(math.asin(sin_alpha))

# --- 8. Solar Azimuth (ψ) ---
num = math.sin(math.radians(H))
den = math.cos(math.radians(H))*math.sin(math.radians(lat)) - math.tan(math.radians(delta))*math.cos(math.radians(lat))
psi = math.degrees(math.atan2(num, den))
if psi < 0:
    psi += 360

# --- 9. Extraterrestrial radiation ---
Gon = Gsc * (1 + 0.033*math.cos(math.radians(360*day/365)))
G0h = Gon * math.sin(math.radians(alpha))

# --- 10. Clear sky radiation ---
Gcs = 0.75 * G0h

# --- 11. Air mass ---
m = 1 / math.sin(math.radians(alpha))

# --- 12. Components ---
Gb = 0.9 * Gcs
Gd = 0.1 * Gcs
Gr = Gcs * albedo * (1 - math.cos(math.radians(lat))) / 2

# --- Print results ---
print(f"--- Douala (4.05°N, 9.7°E) - February 1, 2025 @ 08:00 LT ---")
print(f"Day of Year (n): {day}")
print(f"EoT: {EoT:.2f} min")
print(f"LSTM: {LSTM:.2f}°")
print(f"TC: {TC:.2f} min")
print(f"LST: {LST:.2f} h")
print(f"Hour Angle (H): {H:.2f}°")
print(f"Declination (δ): {delta:.2f}°")
print(f"Altitude (α): {alpha:.2f}°")
print(f"Azimuth (ψ): {psi:.2f}°")
print(f"Extraterrestrial Radiation (G0h): {G0h:.2f} W/m²")
print(f"Clear-sky Radiation (Gcs): {Gcs:.2f} W/m²")
print(f"Air Mass (m): {m:.2f}")
print(f"Direct Beam (Gb): {Gb:.2f} W/m²")
print(f"Diffuse (Gd): {Gd:.2f} W/m²")
print(f"Reflected (Gr): {Gr:.4f} W/m²")
```
